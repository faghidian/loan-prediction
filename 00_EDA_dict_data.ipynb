```python
import pandas as pd
import chardet

# Path to your JSON file
file_path = 'D:/JP Morgan Chase/dict_data.json'

# Detect file encoding
with open(file_path, 'rb') as f:
    result = chardet.detect(f.read())

print(f"File encoding: {result['encoding']}")

# Open the file using the detected encoding
df = pd.read_json(file_path, encoding=result['encoding'])

# Display the DataFrame
df
```
